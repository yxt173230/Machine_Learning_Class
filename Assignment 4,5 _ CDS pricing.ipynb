```python
CRSPdata = pd.read_csv('Quarterly Merged CRSP-Compustat.csv')
CRSPdata.columns = map(str.lower, CRSPdata.columns)
#CRSPdata.columns = [x.lower() for x in CRSPdata.columns]
CRSPdata.head(5)
```

```python
CRSPdata['match_qtr'] = pd.PeriodIndex(pd.to_datetime(CRSPdata.datadate), freq='Q')
cdsdata['match_qtr'] = pd.PeriodIndex(pd.to_datetime(cdsdata.mdate), freq='Q')
```

```python
CRSPdata.head(5)
```

```python
cdsdata['gvkey'] = cdsdata['gvkey'].astype(str).astype(int)
CRSPdata['gvkey'] = CRSPdata['gvkey'].astype(str).astype(int)
com_data = pd.merge(cdsdata, CRSPdata, on=['gvkey','match_qtr'],how='left')
com_data
```
