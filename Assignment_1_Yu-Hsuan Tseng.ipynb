### My process of data cleaning:

```python
# Import data from csv file
import pandas as pd
all_data = pd.read_csv('Assignment_1.csv')
all_data.head(5)
```

```python
# clean the dataframe
all_data = all_data.drop(['datadate', 'gvkey', 'fyearq', 'fqtr', 'indfmt', 'consol', 'popsrc', 'datafmt', 'datafqtr', 'exchg', 'costat' ], axis=1)
all_data = all_data.rename(index=str, columns={'tic':'TIC', 'conm':'C_Name', 'curcdq':'Currency', 'datacqtr':'Qtr', 'atq':'Assets', 'epsfiq':'EPS-IN', 'epsfxq':'EPS-EX', 'lseq':'L+E', 'ltq':'Liabilities', 'niq':'NI_quarter', 'teqq':'Equity', 'xoprq':'Operating_Ex', 'niy':'NI_year', 'revty':'Revenue', 'prccq':'Price_close'})
all_data = all_data[['Qtr','TIC','C_Name','Currency','Assets','Liabilities','Equity','L+E','Revenue','Operating_Ex','NI_quarter','NI_year','Price_close','EPS-IN','EPS-EX']]
all_data = all_data.loc[:, all_data.isnull().mean() < 0.5]
all_data.head(5)
```

```python
# count the missing data
num_missing = pd.DataFrame({'# missing':all_data.isnull().sum()})
num_missing
```

```python
# get the statistics summary
var_stat = all_data.describe()
var_stat.transpose()
```

```python
# combine the the statistics summary with the number of missing data
summary = var_stat.transpose().join(num_missing)
summary = summary.rename(index=str, columns={'count':'# observation', '50%':'median'})
summary = summary[['# observation','# missing','mean','std','25%','median','75%']]
summary
```
